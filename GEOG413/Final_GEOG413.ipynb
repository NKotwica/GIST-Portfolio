```python
import ee
import geemap

map00 = geemap.Map()

data00 = (
    ee.ImageCollection("LANDSAT/LE07/C02/T1_L2")
    .filterDate('2000-01-01', '2000-12-31')
    .filter(ee.Filter.lt('CLOUD_COVER', 50))
)

states00 = ee.FeatureCollection("TIGER/2018/Counties").filter(
    ee.Filter.eq('STATEFP', '12')
)

fl_selected00 = states00.filter(
    ee.Filter.inList('NAME', ['Collier', 'Miami-Dade', 'Broward', 'Monroe'])
)

image00 = (
    data00.median()
    .select(['SR_B1','SR_B2','SR_B3','SR_B4','SR_B5','SR_B7'])
    .multiply(0.0000275)
    .add(-0.2)
)
cropped_image00 = image00.clip(fl_selected00)

everglades00 = geemap.shp_to_ee("EvergladesNP_boundary.shp")
everglades_geom00 = everglades00.geometry()

cropped_masked00 = cropped_image00.updateMask(
    cropped_image00.mask().And(cropped_image00.clip(everglades_geom00).mask().Not())
)

vis_params00 = {
    'bands': ['SR_B3', 'SR_B2', 'SR_B1'],
    'min': 0.0,
    'max': 0.3,
    'gamma': 1.4
}

map00.setCenter(-80.82, 25.49, 8)
map00.addLayer(cropped_masked00, vis_params00, 'Landsat 7 (Everglades Excluded)')
map00.addLayer(everglades00.style(color='gray', fillColor='00000000', width=2), {}, 'Everglades NP Boundary')
map00.addLayer(fl_selected00.style(color='black', fillColor='00000000', width=2), {}, 'Selected Counties')

map00
```

```python
import ee
import geemap

counties = ee.FeatureCollection("TIGER/2018/Counties")

fl_counties = counties.filter(ee.Filter.eq('STATEFP', '12'))

selected_counties = fl_counties.filter(
    ee.Filter.inList('NAME', ['Collier', 'Miami-Dade', 'Broward', 'Monroe'])
)

nlcd_2021 = ee.Image("USGS/NLCD_RELEASES/2019_REL/NLCD/2019").select('landcover')

everglades = geemap.shp_to_ee("EvergladesNP_boundary.shp")
everglades_geom = everglades.geometry()

nlcd_cropped = nlcd_2021.clip(selected_counties)

nlcd_masked = nlcd_cropped.updateMask(
    nlcd_cropped.mask().And(nlcd_cropped.clip(everglades_geom).mask().Not())
)

nlcd_palette = [
    '466b9f', 'd1def8', 'dec5c5', 'd99282', 'eb0000', 'ab0000', 'b3ac9f',
    '68ab5f', '1c5f2c', 'b5c58f', 'ccb879', 'dfdfc2', 'dcd939',
    'ab6c28', 'b8d9eb', '6c9fb8'
]

vis_params = {
    'min': 11,
    'max': 95,
    'palette': nlcd_palette
}

Map = geemap.Map(center=[25.5, -80.8], zoom=8)
Map.addLayer(nlcd_masked, vis_params, 'NLCD 2021 (Everglades Excluded)')
Map.addLayer(selected_counties.style(color='red', fillColor='00000000', width=2), {}, 'Selected Counties')
Map.addLayer(everglades.style(color='green', fillColor='00000000', width=2), {}, 'Everglades NP Boundary')

Map.add_legend(builtin_legend='NLCD')

Map


```
